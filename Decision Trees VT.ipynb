# Importing the libraries

```python
from sklearn import datasets
from random import randrange
import numpy as np 
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor 
```

# Importing the datasets

```python
data1=datasets.load_iris()
data2=datasets.load_breast_cancer()
data3=datasets.load_diabetes()
data4=datasets.load_wine()
data5=datasets.load_digits()
data6=datasets.fetch_california_housing()
data7=datasets.load_boston() 
```

# Creating the dataframe

```python
i=randrange(7)
if(i==0):
    data=data1
    df= pd.DataFrame(data= np.c_[data['data'], data['target']],
                 columns= data['feature_names'] + ['Species'])
elif(i==1):
    data=data2
    df= pd.DataFrame(data= np.c_[data['data'], data['target']],
                 columns= np.append(data['feature_names'],['Target']))
elif(i==2):
    data=data3
    df= pd.DataFrame(data= np.c_[data['data'], data['target']],
                 columns= np.append(data['feature_names'],['Target']))
elif(i==3):
    data=data4
    df= pd.DataFrame(data= np.c_[data['data'], data['target']],
                 columns= np.append(data['feature_names'],['Target']))
elif(i==4):
    data=data5
    df= pd.DataFrame(data= np.c_[data['data'], data['target']],
                 columns=['pixel']*64+['Label'])
elif(i==5):
    data=data6
    df= pd.DataFrame(data= np.c_[data['data'], data['target']],
                 columns= np.append(data['feature_names'],['Target']))
else:
    data=data7
    df= pd.DataFrame(data= np.c_[data['data'], data['target']],
                 columns= np.append(data['feature_names'],['Target']))
```

# Splitting the Dataframe after checking that is it Classification Or Regression

```python
if(df.iloc[:,-1].nunique()<=10):
    X=data.data
    y=data.target
    sc=StandardScaler()
    X=sc.fit_transform(X)
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
    print("Category:Classification")
    print("X_train shape:",X_train.shape,", X_test shape:",X_test.shape,", y_train shape:",y_train.shape,", y_test shape:",y_test.shape)
else:
    X=data.data
    y=data.target
    sc=StandardScaler()
    X=sc.fit_transform(X)
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
    print("Category:Regression")
    print("X_train shape:",X_train.shape,", X_test shape:",X_test.shape,", y_train shape:",y_train.shape,", y_test shape:",y_test.shape)
```

# Applying Decision Tree/Regression Tree Algorithm 

```python
if(df.iloc[:,-1].nunique()<=10):
    clf=DecisionTreeClassifier()
    clf.fit(X_train,y_train)
else:    
    clf=DecisionTreeRegressor()
    clf.fit(X_train,y_train)
```

# Prediction

```python
y_pred=clf.predict(X_test)
```

```python
if(df.iloc[:,-1].nunique()<=10):
    from sklearn.metrics import accuracy_score
    print(accuracy_score(y_test,y_pred))
else:    
    from sklearn.metrics import r2_score
    print(r2_score(y_test,y_pred))
```
